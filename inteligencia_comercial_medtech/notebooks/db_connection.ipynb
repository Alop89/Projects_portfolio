```python
# Librerias 
import pandas as pd
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
# Carga de datos a Neon 

load_dotenv()


file_path = "../data/processed/medtech_sales_data.csv"

db_url = os.getenv("DATABASE_URL")

if not os.path.exists(file_path):
    print(f"Error: El archivo {file_path} no existe")
else: 
    try:
        print(f"Leyende datos desde..{file_path}")

        df_clean = pd.read_csv(file_path)
        if db_url:

            engine = create_engine(db_url)
            table_name = "medtech_sales_data"
            print(f"Iniciando carga a Neon en la tabla {table_name}")
            df_clean.to_sql(table_name, engine, if_exists="replace", index=False)
            print("Carga completada exitosamente")

        else:
            print("Error: No se encontró la variable de entorno DATABASE_URL")
    except Exception as e:
        print(f"Error al cargar los datos a Neon: {e}")
      
    

```
